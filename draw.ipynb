```python
from glob import glob
import numpy as nps
import pickle
from matplotlib import pyplot as plt

data_list = glob("results\exp_2025-04-08-15-*\loss_results.pkl")
# data_list = [
#     r"results\exp_2025-04-07-22-08-26\loss_results.pkl",
#     r"results"
# ]
size = {0:"0.12", 1:"0.06", 2:"0.03", 3:"0.01"}
plt.figure(figsize=(20, 10))
for i, d in enumerate(data_list):
    with open(d, "rb") as f:
        data = pickle.load(f)
    plt.subplot(1, 2, 1)
    # plt.plot(data[0], data[1], alpha=0.3)
    plt.plot(data[3][:400], data[4][:400], label=size[i])
    plt.legend()
    plt.title("Loss")
    plt.subplot(1, 2, 2)
    plt.title("Accuracy")
    plt.plot(data[3][:400], data[5][:400], label=size[i])
    plt.legend()
    
plt.savefig("LearningRate.svg")
```
